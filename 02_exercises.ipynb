# Exploratory Data Analysis
Statistics Canada fuel price dataset — it tracks gasoline prices across Canadian cities over time

- REF_DATE: The month and year of the observation (e.g. Jan-79 = January 1979)
- GEO: The city/region in Canada where the price was recorded
- DGUID: A geographic identifier code used internally by Statistics Canada
- Type of fuel: The fuel type (e.g. regular unleaded gasoline, at full service or self service)
- UOM: Unit of measurement — here it's Cents per litre
- UOM_ID: Numeric code for the unit of measurement
- SCALAR_FACTOR: Scale of the value (here units means no scaling applied)
- SCALAR_ID: Numeric code for the scalar factor
- VECTOR: A unique Statistics Canada series ID for that specific city/fuel combo
- COORDINATE: A numeric code identifying the specific data series combination
- VALUE: The actual fuel price in cents per litre — this is the main column
- STATUS: Quality flag for the data point (NaN = normal, letters indicate issues)
- SYMBOL: Another data quality indicator
- TERMINATED: Whether the data series has been discontinued (t = terminated)
- DECIMALS: Number of decimal places to apply to the VALUE

### Data Wrangling
Data wrangling (sometimes called data munging) is the process of cleaning, transforming, and organizing raw data into a usable format for analysis or modeling.

```python
# select essential columns: REF_DATE, GEO, Type of fuel, Value and save them in new variable
df1=data[["REF_DATE", "GEO", "Type of fuel", "VALUE"]]
df1.head()
```

```python
# change the names-> "REF_DATE" to "DATE", "Type of fuel" : "TYPE"
df1=df1.rename(columns={"REF_DATE":"DATE", "Type of fuel":"TYPE"})
df1.head()
```

```python
# inspect GEO column, you can create two new columns out of it: city and province
# hint: str.split(',',n=1, expand=True)
df1[["City","Province"]]=df1["GEO"].str.split(',',n=1,expand=True)
df1.head()
```

```python
# Use describe() function to get statistical information about the value 
df1.describe()
```

```python
# Check what is inside the categorical variables: use unique, value_counts etc.
print(df1["TYPE"].unique())
print(df1["TYPE"].value_counts())

```

```python
# Select Calgray, Alberta data to see al the infromation
calgary_data=df1[(df1["City"]=="Calgary") & (df1["Province"]==" Alberta")]
print(calgary_data)
```

```python
# 'DATE' is object type, change it to the datetime type to be able to filter by day, month, year etc.
# hint: pd.to_datetime, dt.month_name().str.slice(stop=3), .dt.year
df1["DATE"]=pd.to_datetime(df1["DATE"],format="%b-%y")
df1["Month"]=df1["DATE"].dt.month_name()
df1["YEAR"]=df1["DATE"].dt.year
df1
```

```python
# Select 2000 year
df_2000=df1[df1["YEAR"] ==2000]
df_2000
```

```python
# select the data that shows the price of the 'household heating fuel', in Vancouver, in 1990.
select=df1[(df1["YEAR"]==1990) & (df1["City"]=="Vancouver") & (df1['TYPE']=='Household heating fuel')]
select
```

```python
# Groupby() method's role is to analyze data by some categories
# groupby geo and print number of groups (ngroups)
groups=df1.groupby("GEO")
groups.ngroups
```

Most commonly, we use `groupby()` to split the data into groups,this will apply some function to each of the groups (e.g. mean, median, min, max, count), then combine the results into a data structure. For example, let's select the 'VALUE' column and calculate the mean of the gasoline prices per year. First, we specify the 'Year" column, following by the 'VALUE' column, and the `mean()` function.

```python
# use `groupby()` method to group by the maximum value of gasoline prices, for each month
m =df1.groupby("Month")["VALUE"].max()
m
```
